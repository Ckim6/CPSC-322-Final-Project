## Classification

```python
header, table = myutils.load_csv('input_data_cleaned.csv')
y = []
X = []
for i in table:
    x = [i[1], i[2], i[3], i[4]]
    X.append(x)
    y.append(i[0])
```

```python
# Train folds and test folds for classifiers
n_splits = 10
folds = myevaluation.stratified_kfold_split(X, y, n_splits=n_splits)
```

### Dummy Classifier

```python
# dummy
avg_dummy_accuracy = 0
avg_dummy_error = 0

dummy_matrix_pred = []
dummy_matrix_actual = []
for i in range(n_splits):
    X_train = []
    y_train = []
    X_test = []
    y_test = []
    for j in folds[i][0]:
        X_train.append(X[j])
        y_train.append(y[j])
    for j in folds[i][1]:
        X_test.append(X[j])
        y_test.append(y[j])

    dummy_classifier = MyDummyClassifier()
    dummy_classifier.fit(X_train, y_train)
    predicted_dummy = dummy_classifier.predict(X_test)

    predicted_dummy_instances = []
    for i in predicted_dummy:
        predicted_dummy_instances.append(i[0])

    dummy_accuracy = myevaluation.accuracy_score(y_test, predicted_dummy_instances)
    dummy_error = 1 - dummy_accuracy
    avg_dummy_accuracy += dummy_accuracy
    avg_dummy_error += dummy_error

    dummy_matrix_pred.append(predicted_dummy_instances)
    dummy_matrix_actual.append(y_test)

avg_strat_dummy_accuracy = float("{:.2f}".format(avg_dummy_accuracy / n_splits))
avg_strat_dummy_error = float("{:.2f}".format(avg_dummy_error / n_splits))
```

```python
# calls myclassifiers.py for precision, recall and f1
dummy_precision = myevaluation.binary_precision_score(y_test, predicted_dummy_instances)
dummy_recall = myevaluation.binary_recall_score(y_test, predicted_dummy_instances)
dummy_f1 = myevaluation.binary_f1_score(y_test, predicted_dummy_instances)

# create matrix with totals included
labels = ["MP", "NP"]
knn_matrix = myevaluation.confusion_matrix(y_test, predicted_dummy_instances, labels)
knn_matrix = myutils.matrices_playoff_binary_total(knn_matrix)
```

```python
print("Dummy Classifier (Stratified 10-Fold Cross Validation Results):")
print("--------------------------------------------------------------")
print("Accuracy:", avg_strat_dummy_accuracy)
print("Error Rate:", avg_strat_dummy_error)
print()
print("Precision:", float("{:.2f}".format(dummy_precision)))
print("Recall:", float("{:.2f}".format(dummy_recall)))
print("F1:", float("{:.2f}".format(dummy_f1)))
print()
print("Confusion Matrix")
labels = ["Playoffs", "MP", "NP", "Total"]
print(tabulate(knn_matrix, labels))
```
